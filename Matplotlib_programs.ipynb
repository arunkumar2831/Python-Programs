### Sample Matplotlib Program

Write simple python program display weather chart using Matplotlib.

### M3:Legend with multiple lines

Write a python program to display weather of the day in chart with max,min, avag along with legent to be displayed in plot.

```python
import matplotlib.pyplot as plt # import library
# datas for matplot
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
# chart arguments
plt.xlabel("Days") #title for x axis
plt.ylabel("Temperature") #title for y axis
plt.title("Weather") #title of the chart
plt.plot(days, max_t, label="Max")# label used for legends to discribe the data
plt.plot(days, min_t, label="Min")
plt.plot(days, avg_t, label="Avg")
# plt.legend()
#plt.legend(loc="best", shadow="True", fontsize="large") # default can be decide by chart.
#plt.legend(loc='upper right') # display legend in upper left corner.
plt.legend(loc="best", shadow="True", fontsize="small") # fontsize options are large,small
plt.grid()
```

### M4:Bar Chart.

Write a python program to show the sales of companys using bar chart.

```python
import matplotlib.pyplot as plt
import numpy as np
company=['Tata','Maruti', 'Ashok Leyland,','Mahindra']
revenue=[80,60,50,70]
profit=[50,40,20,35]
#------------------
xpos=np.arange(len(company)) # we use method to show two different charts(revenue, profit) next to each other
print(xpos)
#plt.bar(ypos,revenue)
plt.xticks(xpos,company)
#--------------------------

plt.xlabel("Companys")
plt.ylabel("Revenue")
plt.title("Automobile Company Stocks")
#plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue") # this used for vertical chart "plt.bar"
#plt.bar(xpos+0.2,profit,width=0.4,label="Profit") # in vertical chart we use the width to arrange two bar chart next to each other.
plt.barh(xpos-0.2,revenue,height=0.4,label="Revenue") # this used for horizontal chart "plt.bar"
plt.barh(xpos+0.2,profit,height=0.4,label="Profit")# in horizontal chart we use the height to arrange two bar chart next to each other.
plt.legend()
```

### M5:Histograms Chart

Write a python program to use Histograms chart

```python
import matplotlib.pyplot as plt
blood_sugar_men=[89,95,100,110,90,125,140,150,90,130,80]
blood_sugar_women=[89,105,100,110,90,115,140,150,103,130,80]
#Single histogram chart
#plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,color='g')
#two histogram chart
plt.xlabel("Blood Sugar count")
plt.ylabel("No of patient")
plt.title("Blood Sugar Count Details")
#display multiple values use them as list [men, women]
#bins used as range of values to classify the information.
#rwidth is used display gap between the bins.
#label used to display names.
#orientation is used to display chart in vertical / horizontal.
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=['green','orange'],label=['Men','Women'],
         orientation="horizontal")
plt.legend()
```

### M6:Pie Chart

Write a python program to use Pie chart

```python
import matplotlib.pyplot as plt
exp_val=[1000,600,300,400,250]
exp_label=["House Rent","Food","Mobile Bill","Bike Petrol","Other Expenses"]
plt.axis("equal")#to show perfect circle
plt.pie(exp_val,labels=exp_label,radius=1.2,autopct="%0.2f%%",shadow="True",explode=[0,0.1,0,0,0],startangle=200)
#radius used for set the size of the pie circle by default is 1.
#Autopct is used give percentage and its format.
#explode is used to cut a piece from the pie, used specify which position of chart need to split
#startangle from where pie chart need to start, you need specify the angle start 0 to 360
plt.show()# used to hide unwanted texts
```
