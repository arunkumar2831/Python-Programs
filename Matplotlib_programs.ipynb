### Sample Matplotlib Program

Write simple python program display weather chat using Matplotlib.

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[33,36,40,34,36,38,30]
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather Update')
plt.plot(x,y,color='green', linewidth=4,linestyle='dotted')
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[33,36,40,34,36,38,30]
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather Update')
#this one way of formating the line and plot in the chat.
#plt.plot(x,y,color='green', linewidth=4,linestyle='dotted')
#secord way instead of pass value to arguments we can direct the pass the symbols and format the chat.
#plt.plot(x,y,'rD--') # r - red, D- shape diamond, '--' - line style
#plt.plot(x,y,'m*-.', markersize=10) # we add the markersize as 10.
plt.plot(x,y,'m*-.', markersize=10,alpha=0.7) # here alpha is used for transpence of line, it range from 0 to 1(opacity).
```
