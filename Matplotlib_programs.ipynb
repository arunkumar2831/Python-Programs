### Sample Matplotlib Program

Write simple python program display weather chat using Matplotlib.

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[33,36,40,34,36,38,30]
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather Update')
plt.plot(x,y,color='green', linewidth=4,linestyle='dotted')
```

### M3:Legend with multiple lines

Write a python program to display weather of the day in chat with max,min, avag along with legent to be displayed in plot.

```python
import matplotlib.pyplot as plt # import library
# datas for matplot
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
# chat arguments
plt.xlabel("Days") #title for x axis
plt.ylabel("Temperature") #title for y axis
plt.title("Weather") #title of the chat
plt.plot(days, max_t, label="Max")# label used for legends to discribe the data
plt.plot(days, min_t, label="Min")
plt.plot(days, avg_t, label="Avg")
# plt.legend()
#plt.legend(loc="best", shadow="True", fontsize="large") # default can be decide by chat.
#plt.legend(loc='upper right') # display legend in upper left corner.
plt.legend(loc="best", shadow="True", fontsize="small") # fontsize options are large,small
plt.grid()
```

### M4:Bar Chat.

Write a python program to show the sales of companys using bar chat.

```python
import matplotlib.pyplot as plt
import numpy as np
company=['Tata','Maruti', 'Ashok Leyland,','Mahindra']
revenue=[80,60,50,70]
profit=[50,40,20,35]
#------------------
xpos=np.arange(len(company)) # we use method to show two different chats(revenue, profit) next to each other
print(xpos)
#plt.bar(ypos,revenue)
plt.xticks(xpos,company)
#--------------------------

plt.xlabel("Companys")
plt.ylabel("Revenue")
plt.title("Automobile Company Stocks")
#plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue") # this used for vertical chat "plt.bar"
#plt.bar(xpos+0.2,profit,width=0.4,label="Profit") # in vertical chat we use the width to arrange two bar chat next to each other.
plt.barh(xpos-0.2,revenue,height=0.4,label="Revenue") # this used for horizontal chat "plt.bar"
plt.barh(xpos+0.2,profit,height=0.4,label="Profit")# in horizontal chat we use the height to arrange two bar chat next to each other.
plt.legend()
```
